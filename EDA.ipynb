# Exploratory analysis with wine dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('WineQT.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#univariate analysis,bivariate analysis and multivariate analysis
```

```python
df.describe()
```

```python
df.quality.unique()
```

```python
quality = df.quality.value_counts()
```

```python
df.quality.value_counts().plot(kind='bar',title = 'Quality Wine Distribution')
```

```python
sns.heatmap(df.isnull(),cbar=False,yticklabels=False)
```

```python
df.corr()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=False)
```

```python
plt.figure(figsize=(16,6))
k = 13
cols = df.corr().nlargest(k,'quality')['quality'].index
cm = df[cols].corr()
sns.heatmap(cm,annot=True)
```

```python
values = df.columns.values
number_of_columns = 13
number_of_rows = 13
number_of_rows
```

```python
plt.figure(figsize = (number_of_columns,5*number_of_rows))

for i in range(0,len(values)):
  plt.subplot(number_of_columns,number_of_rows,i+1)
  sns.set_style('whitegrid')
  sns.boxplot(df[values[i]],color='pink',orient='v')
  plt.tight_layout()
```

```python
plt.figure(figsize = (2*number_of_columns,5*number_of_rows))

for i in range(0,len(values)):
  plt.subplot(number_of_rows,number_of_columns,i+1)
  sns.distplot(df[values[i]],kde=True)
```
