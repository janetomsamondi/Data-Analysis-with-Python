**Importing Necessary Libraries**
Pandas for data analysis, Seaborn and Matplotlib for data visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Exploratory Data Analysis


```python
#Loading the datasset
df = pd.read_csv('DP_continuation - Sheet1.csv')
df
```

```python
#check for the dataset shape;number of rows and columns
df.shape
```

```python
#Getting information about the dataset
df.info()
```

```python
df.describe()
```

From the above code, we are able to see possible outliers in Age and Weight.

```python
#checking how many values are actually missing
df.isnull().sum()
```

```python
#checking for unique values
df.nunique()
```

```python
#checking for outliers
df.boxplot(figsize=(20,10))
```

```python
#checking for duplicates
are_there_duplicates = df.duplicated().any()
are_there_duplicates
```

```python
#replacing missing values with median due to presence of outliers
columns_to_fill = ['Weight_' ,'Height']

# Replace missing values with the median for specified columns
for column in columns_to_fill:
    median_value = df[column].median()  # Calculate median for the column
    df[column].fillna(median_value, inplace=True)
```

```python
df.isnull().sum()
```

```python
# key interest in the age variable
plt.boxplot(df['Age'])
plt.xlabel('Age')
plt.ylabel('Values')
plt.title('Box Plot of Data with Outliers')
plt.show()
```

**Data Visulazation**

```python
plt.figure(figsize=(10,10))
sns.barplot(x="Age", y="CountyOfBirth", data =df)
plt.xlabel("Age")
plt.ylabel("CountyOfBirth")
plt.show
```

```python
# Group the data by county of birth and recieve DP
county_counts = df.groupby('CountyOfBirth')['Received_DP'].sum()
county_counts_sorted = county_counts.sort_values(ascending=False)
plt.figure(figsize=(10, 8))
county_counts_sorted.plot(kind='barh', color='skyblue')
plt.xlabel('Sum of Counts of Received DP')
plt.ylabel('County')
plt.title('Sum of Counts of Received DP by County (Sorted)')
plt.show()
```

```python
# Group the data by age and if they  recieve DP
age_counts = df.groupby('Age')['Received_DP'].sum()
age_counts_sorted = county_counts.sort_values(ascending=False)
plt.figure(figsize=(12, 15))
age_counts_sorted.plot(kind='barh', color='skyblue')
plt.xlabel('Sum of Counts of Received DP')
plt.ylabel('Age')
plt.title('Sum of Counts of Received DP by Age (Sorted)')
plt.show()
```

```python
# Group the data by facility type and if they  recieve DP
facility_counts = df.groupby('FacilityType')['Received_DP'].sum()
facility_counts_sorted = facility_counts.sort_values(ascending=False)
plt.figure(figsize=(12, 15))
facility_counts_sorted.plot(kind='barh', color='skyblue')
plt.xlabel('Sum of Counts of Received DP')
plt.ylabel('Facility Type')
plt.title('Sum of Counts of Received DP by Facility (Sorted)')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
for i, month in enumerate(['Refill1month', 'Refill2months', 'Refill3months', 'Refill6months']):
    plt.subplot(1, 4, i+1)
    sns.histplot(df[df[month] == 1]['Age'], kde=True, bins=25, color='skyblue', label='Refill')
    sns.histplot(df[df[month] == 0]['Age'], kde=True, bins=25, color='orange', label='No Refill')
    plt.xlabel('Age')
    plt.ylabel('Frequency')
    plt.title(f'Distribution of Age for Refill {month}')
    plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Calculate the sum of individuals who received DP but did not come for the refill
received_dp_no_refill = df[(df['Received_DP'] == 1) & (df['Refill1month'] == 0)]
sum_received_dp_no_refill = received_dp_no_refill.shape[0]
print(f"{sum_received_dp_no_refill} individuals did not come for refill1month.")

received_dp_no_refill = df[(df['Refill1month'] == 1) & (df['Refill2months'] == 0)]
sum_received_dp_no_refill = received_dp_no_refill.shape[0]
print(f"{sum_received_dp_no_refill} individuals did not come for refill2months.")

received_dp_no_refill = df[(df['Refill2months'] == 1) & (df['Refill3months'] == 0)]
sum_received_dp_no_refill = received_dp_no_refill.shape[0]
print(f"{sum_received_dp_no_refill} individuals did not come for refill3months.")

received_dp_no_refill = df[(df['Refill3months'] == 1) & (df['Refill6months'] == 0)]
sum_received_dp_no_refill = received_dp_no_refill.shape[0]
print(f"{sum_received_dp_no_refill} individuals did not come for refill6months.")
```

```python
# Calculate the number of individuals who received DP
received_dp_counts = df['Received_DP'].value_counts()

# Calculate the number of individuals who came back for refills in each month
refill_counts = df[['Refill1month', 'Refill2months', 'Refill3months', 'Refill6months']].sum()

# Plot the data
plt.figure(figsize=(10, 6))

# Plot for received DP
plt.subplot(1, 2, 1)
received_dp_counts.plot(kind='bar', color='skyblue')
plt.title('Received DP')
plt.xlabel('Received DP')
plt.ylabel('Number of Individuals')

# Plot for refills
plt.subplot(1, 2, 2)
refill_counts.plot(kind='bar', color='orange')
plt.title('Refills')
plt.xlabel('Refill Months')
plt.ylabel('Number of Individuals')

plt.tight_layout()
plt.show()







```

```python
# Combine refill columns into a single column for easier plotting
refill_columns = ['Refill1month', 'Refill2months', 'Refill3months', 'Refill6months']
df['Refills'] = df[refill_columns].sum(axis=1)

# Plot the data
plt.figure(figsize=(12, 6))

# Plot for received DP
plt.subplot(1, 2, 1)
sns.countplot(x='Received_DP', hue='Gender', data=df, palette='viridis')
plt.title('Received DP')
plt.xlabel('Received DP')
plt.ylabel('Number of Individuals')

# Plot for refills
plt.subplot(1, 2, 2)
sns.countplot(x='Refills', hue='Gender', data=df, palette='viridis')
plt.title('Refills')
plt.xlabel('Number of Refills')
plt.ylabel('Number of Individuals')

plt.tight_layout()
plt.show()
```
